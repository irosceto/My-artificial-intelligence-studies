```python
import pandas as pd
```

```python
adv=pd.read_csv('advertising.csv')
```

```python
from scipy import stats
```

```python
linear_regression=stats.linregress(x=adv.TV,
                                  y=adv.Sales)
```

```python
slope=linear_regression.slope
print(slope)
```

```python
intercpt=linear_regression.intercept
print(intercpt)
```

```python
rvalue=linear_regression.rvalue
print(rvalue)
```

```python
tv_spend=200
```

```python
result=slope*tv_spend+intercpt
print(result)
```
